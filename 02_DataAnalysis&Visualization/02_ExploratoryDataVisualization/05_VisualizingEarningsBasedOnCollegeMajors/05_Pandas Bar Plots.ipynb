```python
%matplotlib inline
import pandas as pd
import matplotlib

recent_grads = pd.read_csv('recent-grads.csv')

#Getting warmed up
print(recent_grads['Women'][:5])
recent_grads['Women'][:5].plot(kind='bar')
recent_grads[:5].plot.bar(x='Major', y='Women')
```

```python
recent_grads.sort_values('Median', inplace=True)

# print('Percentages of Women From the 10 Highest Paying Majors')
recent_grads[:5].plot.bar(x='Major', y='ShareWomen')

# print('Percentages of Women From the 10 Lowest Paying Majors')
recent_grads[-5:].plot.bar(x='Major', y='ShareWomen')
```

```python
recent_grads.sort_values('Median', inplace=True)

# print('Percentages of Women From the 10 Highest Unemployment_rate Majors')
recent_grads[:5].plot.bar(x='Major', y='Unemployment_rate')

# print('Percentages of Women From the 10 Lowest Unemployment_rate Majors')
recent_grads[-5:].plot.bar(x='Major', y='Unemployment_rate')
```
