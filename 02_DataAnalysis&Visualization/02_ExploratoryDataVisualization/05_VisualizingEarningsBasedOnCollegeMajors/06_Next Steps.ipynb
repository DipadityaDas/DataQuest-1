```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Use a grouped bar plot to compare the number
# of men with the number of women in each category of majors.
recent_grads = pd.read_csv('recent-grads.csv')

recent_grads.groupby('Major_category')

# Create series with the total Men/Women in
# each major catigories
men_mjr_ctgry = recent_grads.groupby(['Major_category'])['Men'].sum()
women_mjr_ctgry = recent_grads.groupby(['Major_category'])['Women'].sum()

# Determine the number of major categories
N = len(men_mjr_ctgry.values)

fig, ax = plt.subplots(figsize=(10,10))

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

#
rects1 = ax.bar(ind, men_mjr_ctgry.values, width, color='#a4c2f2')
rects2 = ax.bar(ind + width, women_mjr_ctgry.values, width, color='#f2a4d7')

ax.set_xlabel('Major')
ax.set_title('Major Category by gender')
ax.set_xticks(ind + width / 2)

ax.set_xticklabels(men_mjr_ctgry.index, rotation='vertical')

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))

plt.show()
```

```python
# Use a box plot to explore the distributions of median
# salaries and unemployment rate.

fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.boxplot(recent_grads['Median'])
ax2.boxplot(recent_grads['Unemployment_rate'])
ax1.set_xticklabels(['Median Income'])
ax2.set_xticklabels(['Unemployment Rate'])
# ax.set_ylim(0, 5)

plt.show()
```
